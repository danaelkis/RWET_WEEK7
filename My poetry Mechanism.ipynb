### MY POETRY STRUCTURE

Jeopardy! is an American television game show created by Merv Griffin. (Wikipedia). The show features a quiz competition in which contestants are presented with general knowledge clues in the form of answers, and must phrase their responses in the form of questions. 

There are four contestants each show.
The material for the clues covers a wide variety of topics, including history and current events, the sciences, the arts, popular culture, literature, and languages. Category titles often feature puns, wordplay, or shared themes.

In my quest for raw materials, I found a CSV file with 200,000 Jeopardy questions /answers/categories and other information. After reading the file, I found so many beautiful "expressions"  and connection that I could make very fast using the categories names and the answers.

Quick Examples:  <br />

My new poetic form is going to combine answers, categories, and money.

Answers are short and precise, varying from names of movies to names of objects. Since these are answers in a popular trivia show, they all feel familiar.

Categories are surprisingly varied from "stiff" subjects like 'HISTORY' followed by 'BE FRUITFUL & MULTIPLY' or 'NEWS TO ME'.Thinking about it was funny because I would never know what questions can be asked on a theme like 'NEWS TO ME'.

The Amount of money is also important because it will determine the difficulty of the question and will give you an incentive to take a higher risk with the money.  And of course, whether you were correct or wrong. I like that dichotomy in trivia games, there is only one answer and there is only one winner.
